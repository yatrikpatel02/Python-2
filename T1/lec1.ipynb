### pandas series
- pandas series is like a column in a table
- it is a 1-D array holding data of any type

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a,index=["x","y","z"])
print(myvar)
print(type(myvar))
print(myvar[1])                # allowed in older version only
```
