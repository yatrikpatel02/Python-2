# DataFrame

- DataFramee is a 2 dimensional data structure like a 2D array or table with rows and columns

```python
data={'calories':[420,300,400,370],'duration':[50,40,45,35]}
df=pd.DataFrame(data)

print(df)
```

```python
print(df.loc[0])
```

```python
print(df.iloc[0])
```

```python
print(df['calories'])
```

```python
print(df.loc[[0,1]])
```

```python
data={'calories':[420,300,400,370],'duration':[50,40,45,35]}
df=pd.DataFrame(data,index=['day1','day2','day3','day4'])
df
```

```python
print(df.loc['day1'])
```

```python
print(df.iloc['day1'])
```

```python
print(df[::2])
```

```python
data={'Name':['Alice','Bob','Charlie'],'Age':[25,30,35],'city':['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
print('------------------------')
print(df.loc[0:1])
print('------------------------')
print(df.iloc[0:1])
print('------------------------')
print(df.loc[(df.Age>26)])
```
