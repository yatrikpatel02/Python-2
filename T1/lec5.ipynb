#### sort_values()

- sort a dataframe or series by the value along a specific axis

```python
import pandas as pd
data={'Name':['Alice','Bob','Charlie','David'],'Score':[88,92,95,90]}
df=pd.DataFrame(data)
df
```

```python
sorted_df=df.sort_values(by='Score')
sorted_df
```

```python
sorted_df=df.sort_values(by='Score',ascending=False)
sorted_df
```

### apply() 

- used to apply a function along an axis of a dataframe or to the value of series

```python
import pandas as pd
s=pd.Series([35,67,90,45])
def pass_fail(x):
    return 'pass' if x>=50 else 'Fail'
result=s.apply(pass_fail)
print(result)
```

```python
df=pd.DataFrame({
    'physics':[60,70,80],
    'chemistry':[65,75,85],
    'maths':[70,80,90]
})
result=df.apply(sum,axis=0)        # for row summation
print(result)
```

```python
df=pd.DataFrame({
    'physics':[60,70,80],
    'chemistry':[65,75,85],
    'maths':[70,80,90]
})
result=df.apply(sum,axis=1)       # for column summation
print(result)
```

```python
df=pd.DataFrame({
    'physics':[60,70,80],
    'chemistry':[65,75,85],
    'maths':[70,80,90]
})
result=df.apply(lambda x:x.mean())
print(result)
```

```python
def grade(total):
    if total>=250:
        return 'A'
    elif total>=200:
        return 'B'
    else:
        return 'C'
df['total']=df.apply(lambda x:x.sum(),axis=1)
df
```

```python
df['grade']=df['total'].apply(grade)
df
# raw = True(function will give data in numpy array object)
# raw = False(function will give data in form of series)
```

```python
def analyze_score(row):
    total=row.sum()
    status='pass' if total>210 else 'Fail'
    return [total,status]
```

```python
df.apply(analyze_score,axis=1,result_type='expand')
```

```python
df.apply(analyze_score,axis=1,result_type='reduce')
```

```python
df=pd.DataFrame(
{'x':[10,20,30],
 'y':[15,25,35],
 'z':[20,30,40]
})
result=df.apply(lambda x:x-x.mean(),result_type='broadcast')
print(result)
```

## args & kwargs parameter

```python
def add_values(x,y,z):
    return x+y+z
df=pd.DataFrame({
    'A':[1,2],
    'B':[10,20]
})
df
```

```python
df1=df.apply(add_values,args=(1,2))
df1
```

```python
def add_values(x,m):
    return x*m
df=pd.DataFrame({
    'A':[1,2],
    'B':[10,20]
})
df
```

```python
df1=df.apply(add_values,m=10)
df1
```

## concat function 
- used to contibute or stack multiple series or dataframe along a specific axis(row or column)

```python
df1=pd.DataFrame({
    'A':[1,2],
    'B':[3,4]
})
df2=pd.DataFrame({
    'A':[5,6],
    'C':[7,8]
})

result=pd.concat([df1,df2],axis=0)
print(result)
```

```python
df1=pd.DataFrame({
    'A':[1,2],
    'B':[3,4]
})
df2=pd.DataFrame({
    'A':[5,6],
    'C':[7,8]
})

result=pd.concat([df1,df2],axis=1)
print(result)
```

```python
df1=pd.DataFrame({
    'A':[1,2],
    'B':[3,4]
})
df2=pd.DataFrame({
    'A':[5,6],
    'C':[7,8]
})

result=pd.concat([df1,df2],join='outer')
print(result)
```

```python
df1=pd.DataFrame({
    'A':[1,2],
    'B':[3,4]
})
df2=pd.DataFrame({
    'A':[5,6],
    'C':[7,8]
})

result=pd.concat([df1,df2],join='inner')
print(result)
```

```python
df1=pd.DataFrame({
    'A':[1,2],
    'B':[3,4]
})
df2=pd.DataFrame({
    'A':[5,6],
    'C':[7,8]
})

result=pd.concat([df1,df2],join='outer',ignore_index=True)
print(result)
```

```python
df1=pd.DataFrame({
    'E':[1,2],
    'B':[3,4]
})
df2=pd.DataFrame({
    'C':[5,6],
    'D':[7,8]
})

result=pd.concat([df1,df2],sort=False)
print(result)
```

## merge()

- pd.merge(left,right,how='inner',on,left_on,right_on)

```python
# merge()
df=pd.DataFrame({
    'Employee':['A','B','C'],
    'Department':['IT','HR','Finance']
})
df2=pd.DataFrame({
    'Employee':['A','B','D'],
    'Salary':[50000,40000,30000]
})
result=pd.merge(df,df2,on='Employee',how='outer')
result
```

```python
# merge()
df=pd.DataFrame({
    'Employee':['A','B','C'],
    'Department':['IT','HR','Finance']
})
df2=pd.DataFrame({
    'Employee':['A','B','D'],
    'Salary':[50000,40000,30000]
})
result=pd.merge(df,df2,on='Employee',how='inner')
result
```

```python
# merge()
df=pd.DataFrame({
    'Employee':['A','B','C'],
    'Department':['IT','HR','Finance']
})
df2=pd.DataFrame({
    'Employee':['A','B','D'],
    'Salary':[50000,40000,30000]
})
result=pd.merge(df,df2,on='Employee',how='right')
result
```

```python
# merge()
df=pd.DataFrame({
    'Employee':['A','B','C'],
    'Department':['IT','HR','Finance']
})
df2=pd.DataFrame({
    'Employee':['A','B','D'],
    'Salary':[50000,40000,30000]
})
result=pd.merge(df,df2,on='Employee',how='left')
result
```

```python
df1=pd.DataFrame({
    'Emp_ID':[101,102,103],
    'Name':['Amit','Riya','John'],
    'Salary':[50000,60000,55000]
})
df2=pd.DataFrame({
    'Emp_ID':[101,102,104],
    'Name':['Amit','Riya','David'],
    'Salary':[52000,62000,58000]
})
result=pd.merge(df1,df2,on='Emp_ID',suffixes=('_left','_right'))
result
```
