 ### set_index()
- method used to set one or more columns of a dataframe as the index.
- USeful when we need to modify or add row indices to our data as it enhances data retrieved, indexing and merging tasks.

```python
import pandas as pd
data={'Name':['Alice','Bob','Charlie'], 'Age':[25,30,35],'city':['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data)
df
```

```python
df.set_index('Name')
```

```python
df.set_index('Name',drop=False)
```

```python
df.set_index('Name',drop=False,inplace=True)    #inplace update the dataframe and once it is changed then it cannot be retrieved
```

```python
df
```

```python
df.set_index(['Name','Age'],drop=False)    # more than 1 index then use array
```

```python
import pandas as pd
data={'Name':['Alice','Bob','Charlie'], 'Age':[25,30,35],'city':['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data,index=['a','b','c'])
df_reset=df.reset_index(drop=True)                           # reset keeps default index and also add another index
df_reset
```

```python
df.sample()    # gives random sample data
```
