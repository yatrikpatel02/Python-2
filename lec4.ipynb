```python
import requests,json
def get_coordinates():
    api_key="aeb91a913944356cdfa2adfc7b2ce4d0"
    city=input("Enter City Name:")
    request_url=f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response=requests.get(request_url)
    coordinates=response.json()
    print(json.dumps(coordinates,indent=5))
    lattitude=coordinates[0]['lat']
    longitude=coordinates[0]['lon']
    return lattitude,longitude,city
lat,lon,city=get_coordinates()
print(city)
print("--------------------------------------")
print("Lattitude:",lat)
print("Longitude:",lon)
```

```python
import requests,json
def current_weather_data():
    api_key="aeb91a913944356cdfa2adfc7b2ce4d0"
    request_url=f"https://api.openweathermap.org/data/2.5/weather?lat={str(lat)}&lon={str(lon)}&appid={api_key}"
    response=requests.get(request_url)
    current_weather=response.json()
    print(json.dumps(current_weather,indent=5))
    description=current_weather["weather"][0]["description"]
    temp=current_weather["main"]["temp"]
    pressure=current_weather["main"]["pressure"]
    humidity=current_weather["main"]["humidity"]
    wind_speed=current_weather["wind"]["speed"]
    visibility=current_weather["visibility"]
    return description,temp,pressure,humidity,wind_speed,visibility
description,temp,pressure,humidity,wind_speed,visibility=current_weather_data()
print("Description:",description)
print("Temprature:",temp)
print("Pressure:",pressure)
print("Humidity:",humidity)
print("Wind speed:",wind_speed)
print("visibility:",visibility)
```
